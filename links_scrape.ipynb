```python
import requests
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import json
import threading

```

```python
def get_links(start, end, file_num):
    f = open("links/out"+str(file_num)+".txt", "w")
    for i in range(start, end):
        link = "https://housing.com/rent/search-Pskwz0ocdh7q42r5?page="+str(i)
        req = requests.get(link)
        soup = BeautifulSoup(req.text, 'html.parser')
        load = json.loads("".join(soup.find("script", {"type":"application/ld+json"}).contents))
        ls = load[0]['itemListElement']
        for c in ls:
            f.write(c['url'] + '\n')
    f.close()
```

```python
thr = []
n = 50
start = 1
MAX_end = 2527
end = MAX_end//n
inc = MAX_end//n
for i in range(n):
    thr.append(threading.Thread(target = get_links, args = (start, end, i)))
    print(start, end, i)
    start = end
    end = end + inc

```

```python
for t in thr:
    t.start()
```

```python
def cat(n):
    f = open("links/out.txt", "w")
    for i in range(0, n+1):
        g = open("links/out"+str(i)+".txt", "r")
        txt = g.read()
        f.write(txt)
        print(i, len(txt.rsplit('\n')))
        g.close()
    f.close()
```

```python
def fetch_data(link):
    req = requests.get(link)
    soup = BeautifulSoup(req.text, 'html.parser')
    if req.status_code != 200:
        print("ERROR:" + link)
    item = {}
    item["link"] = link    
    
    load = json.loads("".join(soup.find("script", {"type":"application/ld+json"}).contents))
    load[0]["geo"]["latitude"]
    item["latitude"] = load[0]["geo"]["latitude"]
    item["longitude"] = load[0]["geo"]["longitude"]
    
    
    overview = soup.find("tbody", {"class": "css-twrsy5"})
    for row in overview:
        item[row.th.text] = row.td.text
    print(item)
```
