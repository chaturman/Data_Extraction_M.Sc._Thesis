```python
def isNaN(num):
    return num != num
```

```python
def process_buildup(buildup):
    if buildup =="" or isNaN(buildup):
        return -1
    if isinstance(buildup,int) or isinstance(buildup,float):
        return buildup
    return int(buildup.split(" ")[0])
```

```python
def process_added(added):
    if added=="" or isNaN(added):
        return 0
    if added=="Just now":
        return 0
    if added == "More than a month ago":
        return 100
    else:
        return int(added.split(" ")[0])
```

```python
def process_carpet(carpet):
    if carpet =="":
        return -1
    if isinstance(carpet,int) or isinstance(carpet,float):
        return carpet
    return int(carpet.split(" ")[0])
```

```python
def process_age(age):
    if age=="":
        return 0
    if isinstance(age, int) or isinstance(age,float):
        return age
    else:
        return int(age.split(" ")[0])
```

```python
genericlist= ['Bore Water', 'Bore and Municipal water', 'Garden',
       'Gated Community', 'Gym', 'Intercom', 'Kids Area', 'Lift',
       'Municipal Water', 'Parking', 'Pet Allowed', 'Pool', 'Security',
       'Sports Facility', 'Water Supply']
```

```python
# Lift:0, Swimming pool:1, Gym:2, Parking:3, Gated Community:4
def process_amen(amen):
    amen_list = []
    for i in genericlist:
        if i in amen:
            amen_list.append(1)
        else:
            amen_list.append(0)
    return amen_list
```

```python
# gaspipline:0, powerbackup:1
soc_list = ['AC', 'Bed', 'Cupboard', 'Fridge', 'Gas Pipeline', 'Geyser', 'Intercom', 'Microwave', 'Power Backup', 'Sofa', 'Stove', 'TV','Washing Machine']
def process_furn(furn):
    list_furn = []
    for i in soc_list:
        if i in furn:
            list_furn.append(0)
        else:
            list_furn.append(1)
    return list_furn
```

```python
def process_balcony(balcony):
    if(balcony==""):
        return 0
    if(balcony=="No Balcony" or isNaN(balcony)):
        return 0
    else:
        return int(balcony)
```

```python
def process_parking(parking):
    if(parking==""):
        return 0
    if(parking=="No Parking"):
        return 0
    else:
        if isinstance(parking,int) or isinstance(parking,float):
            return parking
        return int(parking.split(" ")[0])
```

```python
# 1 for lower, 2 for middle, 3 for higher
def process_floorinfo1(f1):
    if isNaN(f1):
        return 0
    f1 = f1.lower()
    if(f1==""):
        return 0
    if("lower" in f1 or "ground floor" in f1 or "1st" in f1):
        return 1
    elif("middle" in f1):
        return 2
    elif "higher" in f1:
        return 3
    else:
        return 0
```

```python
def process_floorinfo2(f2):
    if(isinstance(f2,int) or isinstance(f2,float)):
        return f2
    if f2=="" or len(f2.split(" "))<3:
        return 0    
    return int(f2.split(" ")[2])
```

```python
# 0 for unfurnished, 1 for semi furnished, 2 for fully furnished
def process_furnishing(fur):
    if(fur == "Unfurnished"):
        return 0
    elif(fur=="Semi Furnished"):
        return 1
    elif fur=="Fully Furnished":
        return 2
    else:
        return 0
```

```python
def process_facing(facing):
    if isinstance(facing,float):
        return facing
    facing = facing.lower()
    if facing=="north":
        return 0
    elif facing=="north-east":
        return 1
    elif facing=="east":
        return 2
    elif facing=="south-east":
        return 3
    elif facing=="south":
        return 4
    elif facing=="south-west":
        return 5
    elif facing=="west":
        return 6
    elif facing=="north-west":
        return 7
    else:
        return -1
```

```python
def process_price(price):
    if(price=="" or isNaN(price)):
        return 0
    num1 = (price.split(" ")[0])
    numeric_string = re.sub("[^0-9.]", "", num1)
    num1 = float(numeric_string)
#     if len(price.split(" "))>1:      
#         if price.split(" ")[1]=="L":
#             return num1 * 100000
#         else:
#             return num1 * 10000000
#     else:
#         return 0
    return num1
```

```python
def process_emi(emi):
    if(emi==""):
        return 0
    num1 = float(emi.split(" ")[0])
    if len(emi.split(" "))>1:      
        if emi.split(" ")[1]=="K":
            return num1 * 1000
        else:
            return num1 * 100000
    else:
        return 0
```

```python
monthmap = {
    "Jan,": "01",
    "Feb,": "02",
    "Mar,": "03",
    "Apr,": "04",
    "May,": "05",
    "Jun,": "06",
    "Jul,": "07",
    "Aug,": "08",
    "Sep,": "09",
    "Oct,": "10",
    "Nov,": "11",
    "Dec,": "12"
}
```

```python
def process_possession(poss):
    if poss=="":
        return 0
    if poss=="Available now":
        return 0
    else:
        month = ""
        if isinstance(poss,float) or isinstance(poss,int):
            return 0
        arr = poss.split(" ")
        if(len(arr)!=2):
            return 0
        year = arr[1]
        for mon in monthmap:
            if mon in poss:
                month = mon
        if month == "":
            return 0
        ans = monthmap[month] + "-" + year
        return ans
```

```python
import re
def process_brokerage(brok):
    if brok=="":
        return 0
    if isinstance(brok, float) or isinstance(brok, int):
        return brok
    if "No Charge" in brok:
        return 0
    else:
        numeric_string = re.sub("[^0-9.]", "", brok)
        num1 = float(numeric_string)
        if("Lac" in brok):
            return num1 * 100000
        if("Cr" in brok):
            return num1 * 10000000
        return num1
```

```python
def process(df):
    for i, row in df.iterrows():
        df.at[i, 'Age of property'] = process_age(row['Age of property'])
        #df.at[i, 'Avg. Price'] = process_avg(row['Avg. Price'])
        df.at[i, 'Balcony'] = process_balcony(row['Balcony'])
        df.at[i,'Bedrooms'] = (row['Bedrooms'])
        df.at[i,'Brokerage'] = process_brokerage(row['Brokerage'])
        df.at[i,'Build Up Area'] = process_buildup(row['Build Up Area'])
        df.at[i,'Carpet Area'] = process_carpet(row['Carpet Area'])
        df.at[i,'Furnishing'] = process_furnishing(row['Furnishing'])
        df.at[i,'Parking'] = process_parking(row['Parking'])
        df.at[i,'Possession status'] = process_possession(row['Possession status'])
        df.at[i,'Price'] = process_price(row['Price'])
        #df.at[i,'emi'] = process_emi(row['emi'])
        df.at[i,'floorinfo1'] = process_floorinfo1(row['Floor number'])
        df.at[i,'floorinfo2'] = process_floorinfo2(row['Floor number'])
        
        df.at[i,'Gate Community'] = process_gated(row['Gate Community'])
        df.at[i,'Security'] = process_sec(row['Security'])
        df.at[i,'Gas Pipeline'] = process_gaspipe(row['Gas Pipeline'])
        
        df.at[i,'amen'] = process_amen(row['amen'])
        df.at[i,'furn'] = process_furn(row['furn'])
        df.at[i,'Facing'] = process_facing(row['Facing'])
        df.at[i,'Age of property'] = process_age(row['Age of property'])
#         df.at[i,'Added'] = process_added(row['Added'])
        
        if( i%5000 == 0):
            print(i)
```
